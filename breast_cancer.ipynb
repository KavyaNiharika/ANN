# Categorical Value

```python
df['diagnosis'].value_counts()
```

```python
df['diagnosis']=df['diagnosis'].map({'B':0,'M':1}).astype(int)
df.head()
```

```python
x=df.drop(['diagnosis'],axis=1)
y=df['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

# Scaling Down

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
xtrain=sc.fit_transform(xtrain)
xtest=sc.transform(xtest)
```

```python
xtrain.shape
```

# ANN

```python
import keras
from keras.models import Sequential
from keras.layers import Dense,Dropout,BatchNormalization,Activation,Embedding,Flatten
from keras.activations import relu,sigmoid
from keras.layers import ReLU,LeakyReLU
```

```python
ann=Sequential()
```

```python
ann.add(Dense(units=25,kernel_initializer='he_uniform',activation='relu',input_dim=30))
ann.add(Dense(units=20,kernel_initializer='he_uniform',activation='relu'))
ann.add(Dense(units=10,kernel_initializer='he_uniform',activation='relu'))
ann.add(Dense(units=6,kernel_initializer='he_uniform',activation='relu'))
ann.add(Dense(units=1,kernel_initializer='glorot_uniform',activation='sigmoid'))
```

```python
ann.compile(optimizer='AdaDelta',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
reg=ann.fit(xtrain,ytrain,validation_split=0.0,batch_size=10,epochs=100)
```

```python
ypred=ann.predict(xtest)
ypred=ypred>0.5
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
accuracy_score(ytest,ypred)
```

```python
print(confusion_matrix(ytest,ypred))
```
